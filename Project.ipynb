```python
# the attributes are not in normal form to be understanable

# So, let's rename columns as per the dataset descrtion
colNames = ["erythema" , "scaling", "definite_borders", "itching", "koebner_phenomenon", "polygonal_papules", "follicular_papules", "oral_mucosal_involvement", "knee and elbow involvement", "scalp_involvement", "family_history",
"melanin_incontinence","eosinophils_in_the_infiltrate", "PNL_infiltrate", "fibrosis_of_the_papillary_dermis", "exocytosis", "acanthosis", "hyperkeratosis", "parakeratosis", "clubbing_of_the_rete_ridges", "elongation_of_the_rete_ridges", "thinning_of_the_suprapapillary_epidermis", "spongiform_pustule", "munro_microabcess", "focal_hypergranulosis", "disappearance_of_the_granular_layer", "vacuolisation_and_damage_of_basal_layer", "spongiosis", "saw-tooth_appearance_of_retes", "follicular_horn_plug", "perifollicular_parakeratosis", "inflammatory_monoluclear_inflitrate", "band-like_infiltrate", "Age", "Disease"]


# changing column names into readable foramt as per the dataset description
# dataset.columns = colNames

```

```python
# renaming columns and replacing null values '?' with np.nan
dataset = pd.read_csv(url, names=colNames, na_values='?')
```

```python
# method to overcome missing values
# dataset["Age"].fillna( method ='ffill', inplace = True)  # 85% accuracy
dataset['Age']=dataset['Age'].fillna(dataset['Age'].median())
```

```python
# since null values '?' was there previosuly, the data type was of Float type
dataset['Age'] = np.int64(dataset['Age'])

# Now, typecasting back to int type for performing operations ahead else, error will be raised
```

```python
# data wrangling and mining done

# Preparing our dataset for input and target values

array = dataset.values  # change dataframe
X = array[:, 0:-1] # all rows taken, all columns except last one
Y = array[:, -1] # all rows takne, only last column taken
```

```python
from sklearn.model_selection import train_test_split
# split dataset into train , test and validation sets
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=7)

# 80% training
# 20% testing
```

```python
# importing sklearn classification models

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# K Neighbors classifier
clf=KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
#print(y_pred)

print(clf.score(x_test, y_test))
print('misclassified samples:%d'%(y_test!=y_pred).sum())
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('Accuracy score %.2f'%accuracy_score(y_test,y_pred))
```

```python
# Standaradising dataset for better training of dataset

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
# K Neighbors classifier
clf=KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train_sc,y_train)
y_pred=clf.predict(x_test_sc)
#print(y_pred)

print(clf.score(x_test_sc, y_test))
print('misclassified samples:%d'%(y_test!=y_pred).sum())
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('Accuracy score %.2f'%accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification

from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

from sklearn.metrics import precision_score, recall_score, auc, roc_curve, f1_score

import warnings
warnings.filterwarnings('ignore')
```

```python
names = [
    'Nearest Neighbors', 'Linear SVM', 'RBF SVM', 'Gaussian Process', 'Decision Tree', 'Random Forest', 'Neural Network', 'AdaBoost', 'Naive Bayes', 'QDA'
]

classifiers = [
KNeighborsClassifier(3),
SVC(kernel='linear', C=0.025),
SVC(gamma=2, C=1),
GaussianProcessClassifier(1.0 * RBF(1.0)),
DecisionTreeClassifier(max_depth=6),
RandomForestClassifier(max_depth=6, n_estimators=10, max_features=1),
MLPClassifier(alpha=1, max_iter=500),
AdaBoostClassifier(),
GaussianNB(),
QuadraticDiscriminantAnalysis()

]
```

```python
model_cols=[]
df = pd.DataFrame(columns=model_cols)
index=0
```

```python
model_cols=[]
df = pd.DataFrame(columns=model_cols)
index=0
for name, clfr in zip(names, classifiers):
    clfr.fit(x_train, y_train)
    df.loc[index, 'Classifiers'] = name
    df.loc[index, 'Train Accuracy'] = clfr.score(x_train, y_train)
    df.loc[index, 'Test Accuracy'] = clfr.score(x_test, y_test)
    df.loc[index, 'Precision'] = precision_score(y_test, clfr.predict(x_test), average='macro')
    df.loc[index, 'Recall'] = recall_score(y_test, clfr.predict(x_test), average='macro')

    df.loc[index, 'F1 Score'] = f1_score(y_test, clfr.predict(x_test), average='macro')


    index += 1

```
